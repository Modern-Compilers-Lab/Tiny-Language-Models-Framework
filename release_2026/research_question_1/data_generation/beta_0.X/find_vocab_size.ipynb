```python
import json

# Open and parse the JSON file
with open("bpe_tokenizer.json", "r", encoding="utf-8") as f:
    tokenizer_data = json.load(f)

# The vocabulary is usually nested inside the 'model' key
try:
    vocab_size = len(tokenizer_data["model"]["vocab"])
    print(f"BPE Vocab Size: {vocab_size}")
except KeyError:
    print("Could not find the 'model' -> 'vocab' keys. Print tokenizer_data.keys() to check the structure.")
```

```python
# Open the file and count the lines
with open("sp_unigram_newest.vocab", "r", encoding="utf-8") as f:
    vocab_size = sum(1 for line in f)

print(f"SP Unigram Vocab Size: {vocab_size}")
```
